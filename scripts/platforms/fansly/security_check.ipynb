```typescript
const Security = (function () {
  function imul(a, b) {
    const ah = (a >>> 16) & 0xffff
    const al = a & 0xffff
    const bh = (b >>> 16) & 0xffff
    const bl = b & 0xffff
    return (al * bl + (((ah * bl + al * bh) << 16) >>> 0)) | 0
  }

  // Fonction de hachage cyrb53 (version exacte du code que tu as fourni)
  function cyrb53(q, seed = 0) {
    let N = 3735928559 ^ seed,
      ne = 1103547991 ^ seed
    for (let _e, oe = 0; oe < q.length; oe++) {
      _e = q.charCodeAt(oe)
      N = imul(N ^ _e, 2654435761)
      ne = imul(ne ^ _e, 1597334677)
    }
    N = imul(N ^ (N >>> 16), 2246822507)
    N ^= imul(ne ^ (ne >>> 13), 3266489909)
    ne = imul(ne ^ (ne >>> 16), 2246822507)
    ne ^= imul(N ^ (N >>> 13), 3266489909)
    return 4294967296 * (2097151 & ne) + (N >>> 0)
  }

  function publicGenerateRequestCheck(url: string, deviceId: string) {
    // Générer la checkKey comme dans le code trouvé
    const checkKey = ['necvac', 'govry3', 'tybkYz'].join('-')

    // Extraire le pathname de l'URL
    const pathname = new URL(url).pathname

    // Créer la chaîne à hacher (concaténation)
    const hashInput = `${checkKey}_${pathname}_${deviceId}`

    // Générer le hash en utilisant la fonction cyrb53
    const hash = cyrb53(hashInput)

    // Convertir le hash en hexadécimal
    const hashHex = hash.toString(16)

    return hashHex
  }

  return {
    GenerateRequestCheck: publicGenerateRequestCheck
  }
})()

```
